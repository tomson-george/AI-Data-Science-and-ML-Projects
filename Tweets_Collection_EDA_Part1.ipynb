# 1. Pulling Tweets

```python
#importing the necessary libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import re
import string
import datetime

import tweepy as tw
import nltk


ACCESS_TOKEN = '##Access Token##'
ACCESS_TOKEN_SECRET = '##Access Token Secret##'

CONSUMER_API_KEY = '##Consumer API Key##'
CONSUMER_API_SECRET = '##Consumer API Secret##'

#Create an authentication object
my_auth_obj = tw.OAuthHandler(CONSUMER_API_KEY, CONSUMER_API_SECRET)

#Set access token to auth object
my_auth_obj.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

#Create an API Object to get tweets using authentication object
my_api = tw.API(my_auth_obj)
filetags = ["Altcoin", "Bitcoin", "Coindesk", "Cryptocurrency", "Gold", "APPL", "GOOG", "YHOO"]
```

## Creating the files - ONE TIME EXECUTION

```python
#Creating files for all tags(ONE TIME EXECUTION)
for item in filetags:
    filepath = 'data/' + item + '.csv'
    df = pd.DataFrame(columns=['Tweet ID', 'Time of Tweet', 'User ID','Text'])
    df.to_csv(filepath, mode='a', index=False, header=True)
```

## Defining functions for scrapping tweets and writing them to the file

```python

#Function to scrap data
def scrap(date):
    query = "#Altcoin OR #Bitcoin OR #Coindesk OR #Cryptocurrency OR #Gold OR #APPL OR #GOOG OR #YHOO"
    tweets = tw.Cursor(my_api.search_tweets, q = query, until = date, tweet_mode = 'extended').items(2000)
    print(f"Scrapping completed Successfully for {date}!")
    return tweets

#Function for writing pulled data into dataframe and file
def write_to_df_and_file(word,tweetsdf):
    filepath = 'data/' + word + '.csv'
    
    #Getting length of the existing data
    df = pd.read_csv(filepath)
    l = df.index
    #Dropping all the rows in dataframe
    df.drop(l, inplace=True)
    
    hashw = '#'+word
    
    for i in range(len(tweetsdf)):
        if re.search(hashw.lower(),tweetsdf.loc[i, "Text"].lower()):
            df.loc[len(df.index)] = [tweetsdf.loc[i,"Tweet ID"], tweetsdf.loc[i,"Time of Tweet"], tweetsdf.loc[i,"User ID"], tweetsdf.loc[i,"Text"]]

    df.to_csv(filepath, mode='a', index=False, header=False)
    print(f"Tweet data about '{word}' is written into corresponding file.")
```

## Pulling tweets date wise for all tags and saving them in file

```python
#Pulling data before 2022-11-06 00:00:00
hashtags = ["Altcoin", "Bitcoin", "Coindesk", "Cryptocurrency", "Gold", "APPL", "GOOG", "YHOO"]
date = '2022-11-06'
tweets = scrap(date)

#Main dataframe for storing tweets
fdf = pd.DataFrame(columns=['Tweet ID', 'Time of Tweet', 'User ID','Text'])

for tweet in tweets:
    tweet_id = tweet.id
    time_of_tweet = tweet.created_at
    user_id = tweet.user.id
    text = tweet.full_text
    fdf.loc[len(fdf.index)] = [tweet_id, time_of_tweet, user_id, text]
    
print(f"Data is written to the main dataframe of shape {fdf.shape}.")

for tag in hashtags:
    write_to_df_and_file(tag,fdf)
    
#Dropping main dataframe
del fdf
print(f"Droped the main dataframe of date {date}!")
```

# 2. Clening the dataset

```python
#Initialising combined data frame
hashtags = ["Altcoin", "Bitcoin", "Coindesk", "Cryptocurrency", "Gold", "APPL", "GOOG", "YHOO"]
cdf = pd.DataFrame(columns=['Tweet ID', 'Time of Tweet', 'User ID','Text','Tag'])

#Taking data from each file and appending to the maindataframe, and its type
for tag in hashtags:
    df = pd.read_csv('data/'+tag+'.csv')
    df["Tag"] = [tag] * df.shape[0]
    cdf = pd.concat([cdf,df])
```

## Dropping duplicates

```python
#Size before dropping duplicates
print(cdf.shape)

#dropping duplicates from combined dataframe
cdf.drop_duplicates()

#Size after dropping duplicates
print(cdf.shape)
```

## Remove Punctuations

```python
#Remove Punctuations
cdf["Text"] = cdf["Text"].map(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
cdf.head(10)
```

## Removing Emojis

```python
#Removing Emojis
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)


cdf["Text"] = cdf["Text"].map(lambda x: remove_emoji(x))
cdf.head(10)
```

## Removing Numbers

```python
#Removing numbers in Tweet
def remove_numbers(string):
    pattern = re.compile("[0-9]+")
    return pattern.sub(r'', string)

cdf["Text"] = cdf["Text"].map(lambda x: remove_numbers(x))
cdf.head(20)
```

## Removing words having length less than 2

```python
#Remove words with length less than 2
def remove_words_lessthan_two(string):
    pattern = re.compile(r'\b\w{1}\b')
    return pattern.sub(r'',string)

cdf["Text"] = cdf["Text"].map(lambda x: remove_words_lessthan_two(x))
cdf.head(20)
```
